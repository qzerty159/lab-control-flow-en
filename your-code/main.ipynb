## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
for word1 in words :
    print(f"The word {word1} in upper case is {word1.upper()}")
```

**Create a new list containing only words with 5 or more letters**

```python
words_five_letters_or_more = []

for word2 in words :
    if len(word2) >= 5  :
        words_five_letters_or_more.append(word2)

print(f"The new list containing only words with 5 or more letters is : \n\nwords_five_letters_or_more = {words_five_letters_or_more}")
words_five_letters_or_more[0]
```

**Print the first word starting with "t"**

```python
for word3 in words_five_letters_or_more :
    for i in word3 :
        if i[0] == "t" :
            print(f"The first word starting with \"t\" is the word \"{word3}\"")
        else : 
            break
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squares_every_number = []

for num1 in range(1, 11):
    squares_every_number.append(num1**2)

print(f"The list containing the square of every number from 1 to 10 is : \n\nsquares_every_number = {squares_every_number}")
```

**Print a list containing the square of every odd number from 1 to 10**

```python
squares_odd_number = []

for num2 in range(1, 11) :
    if num2 % 2 != 0 :
        squares_odd_number.append(num2**2)

print(f"The list containing the square of every odd number from 1 to 10 is : \n\nsquares_odd_number = {squares_odd_number}")
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
squares_multiples_of_8 = []

for num3 in range(1000) :
    if num3 % 8 == 0 :
        squares_multiples_of_8.append(num3**2)

print(f"The list containing the square of all multiples of 8 below 1000  is : \n\nsquares_odd_number = {squares_multiples_of_8}")
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
count1 = 0

for name1 in people :
    count1 += 1
    
print(f"They are {count1} peoples")
```

**How many people have kids**?

```python
count2 = 0

for name2 in people :
    if name2["n_kids"] > 0 :
            count2 += 1

print(f"They are {count2} peoples who have kids")
```

**How many kids do they have in total?**

```python
count3 = 0

for name3 in people :
    if name3["n_kids"] > 0 :
        count3 += name3["n_kids"]

print(f"Peoples have a total of {count3} kids in total")
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
people_extrat_kid = []

for person in people :
    new_person = person.copy()
    if new_person["name"].endswith("a"):
        new_person["n_kids"] += 1
        people_extrat_kid.append(new_person)

print(people_extrat_kid)
```
